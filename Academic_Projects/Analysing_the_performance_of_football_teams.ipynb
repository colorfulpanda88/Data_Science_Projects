# Read the data set

```python
import pandas as pd
```

```python
la_liga = pd.read_csv('Laliga.csv')
```

```python
la_liga.head(10)
```

```python
la_liga.columns = la_liga.iloc[0]
la_liga.drop([0], axis=0, inplace=True)
la_liga.replace(to_replace='-', value='0', inplace=True)
```

```python
la_liga.head(10)
```

# Teams which have started playing between 1930-1980.

```python
df = la_liga['Debut'].str.slice(0, 4).astype(int)
la_liga_modified = la_liga[(df > 1929) & (df < 1981)]
la_liga_modified[['Team', 'Debut']]
```

# Print the list of teams which came Top 5 in terms of points

```python
la_liga['Points'] = la_liga['Points'].apply(pd.to_numeric) 
sorted_la_liga = la_liga.sort_values(by='Points', axis=0, ascending=False)
sorted_la_liga[['Team', 'Points']].head(5)
```

# Finding the team which has maximum and minimum goal difference.

```python
import numpy as np

def Goal_diff_count():
    return  la_liga['GoalsFor'] - la_liga['GoalsAgainst']

la_liga['GoalsFor'] = la_liga['GoalsFor'].apply(pd.to_numeric) 
la_liga['GoalsAgainst'] = la_liga['GoalsAgainst'].apply(pd.to_numeric)
la_liga['Goals_Difference'] = Goal_diff_count()

max_location = la_liga.index.get_loc(la_liga['Goals_Difference'].idxmax())
min_location = la_liga.index.get_loc(la_liga['Goals_Difference'].idxmin())
 
max_goal_difference_team = la_liga.iloc[max_location, 1]
min_goal_difference_team = la_liga.iloc[min_location, 1]

print ("Team ", max_goal_difference_team, " has maximum goal difference ", la_liga['Goals_Difference'].max())
print ("Team ", min_goal_difference_team, " has minimum goal difference ", la_liga['Goals_Difference'].min())
```

# Creating a new column with the name “Winning Percent"
## Percentage of Winning = (GamesWon / GamesPlayed) * 100
## If there are any numerical error, replacing with 0%, also printing the top 5 teams which have the highest winning percentage

```python
la_liga['GamesWon'] = la_liga['GamesWon'].apply(pd.to_numeric) 
la_liga['GamesPlayed'] = la_liga['GamesPlayed'].apply(pd.to_numeric)
Percentage_of_Winning = (la_liga['GamesWon'] / la_liga['GamesPlayed'])*100
la_liga['Percentage_of_Winning'] = Percentage_of_Winning

la_liga_top_5 =la_liga.sort_values('Percentage_of_Winning',ascending = False)
la_liga_top_5[['Team', 'Percentage_of_Winning']].head(5)
```

# Group teams based on their “Best position” and print the sum of their points for all positions

```python
group_la_liga_data = la_liga.groupby(['BestPosition'])['Points'].sum()
```
